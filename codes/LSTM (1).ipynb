```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Define initial positions for device and anchor nodes
deviceLocInitial = np.array([50, 50])  # Initial device position
nodeLocInitial = np.array([[40, 41],     # Initial anchor node positions
                           [62, 83],
                           [87, 24]])

# Define velocity for device (change as needed)
deviceVelocity = np.array([0.5, 0.2])   # Velocity of the device in x and y directions

# Number of anchor nodes
numNodes = nodeLocInitial.shape[0]

# Simulation parameters
numTimeSteps = 100  # Number of time steps for simulation
timeStep = 0.1      # Time step duration (adjust as needed)
SNR = 10            # Signal-to-Noise Ratio

# Initialize device and anchor node positions
deviceLoc = deviceLocInitial.copy()
nodeLoc = nodeLocInitial.copy()

# Initialize LSTM model
model = Sequential()
model.add(LSTM(units=50, activation='relu', input_shape=(1, 2)))
model.add(Dense(units=2))
model.compile(optimizer='adam', loss='mse')

# Prepare data for LSTM training
X, y = [], []
for t in range(numTimeSteps):
    X.append(deviceLoc.copy())
    next_positions = deviceLoc + deviceVelocity * timeStep
    y.append(next_positions)
    deviceLoc = next_positions.copy()

X = np.array(X)
y = np.array(y)

# Reshape data for LSTM input
X = X.reshape(X.shape[0], 1, X.shape[1])

# Train LSTM model
model.fit(X, y, epochs=100, verbose=0)

# Predict trajectory using LSTM
predicted_positions = []
deviceLoc = deviceLocInitial.copy()
for t in range(numTimeSteps):
    X_test = np.array([deviceLoc])
    X_test = X_test.reshape(1, 1, 2)
    predicted_position = model.predict(X_test)
    predicted_positions.append(predicted_position[0])
    deviceLoc = predicted_position[0]

predicted_positions = np.array(predicted_positions)

# Plotting
plt.figure()
plt.plot(y[:, 0], y[:, 1], 'ro-', label='Actual Trajectory')
plt.plot(predicted_positions[:, 0], predicted_positions[:, 1], 'bx-', label='Predicted Trajectory')
plt.xlabel('X Position')
plt.ylabel('Y Position')
plt.title('Dynamic UWB Localization')
plt.legend()
plt.show()
```
